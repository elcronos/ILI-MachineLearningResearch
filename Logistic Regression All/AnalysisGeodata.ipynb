```python
def words_in_string(word_list, a_string):
    return set(word_list).intersection(a_string.split())

def string_in_another(word_list, location):
    for word in word_list:
        if location.lower().find(word) != -1:
             return True
    return False

# Remove some characters
def remove_special_characters(text):
    text = text.replace(',',' ')
    text = text.replace('/',' ')
    bad_chars = '#?(){}<>:;.!$%&/=+*^-'
    rgx = re.compile('[%s]' % bad_chars)
    return rgx.sub('', text)

QUEENSLAND = ['brissy','brisbane','qld','queensland','seq','gold coast','caloundra','townsville','sunshine coast','nambour']
NSW= ['new south wales', 'nsw', 'newcastle','sydney', 'blue mountains','lithgow','campbelltown','bondi beach','edenmonaro','nowra','[s]ydney','syd']
SA = ['south australia','adelaide','sa','gawler','kaurna']
ACT= ['capital','act','camberra','canberra']
TASMANIA = ['tassy','tasmania','hobart','launceston']
WA = ['south hedland', 'port hedland','nyabing','wa','western australia', 'perth','fremantle','freo']
VICTORIA = ['melbourne', 'victoria','dandenong','melb']
NT = ['darwin','nt','northern territory']

#Bounding Boxes
BB_QLD=[138.43,-28.19,153.54,-10.79]
BB_NSW1=[141.02, -34.04,154.12, -29.09]
BB_NSW2=[148.27,-36.97,153.76,-29.02]
BB_SA=[129.07,-35.73,140.67,-26.02]
BB_VIC=[139.46,-38.93,150.25,-36.24]
BB_WA=[113.38, -35.28, 129.11, -12.25]
BB_TAS=[143.88,-44.12,149.68,-39.54]
BB_ACT=[148.7714,-35.9169,149.3921,-35.0941]
BB_NT=[129.16,-25.86,137.81,-10.68]


def find_location(lati, longi):
    if(BB_ACT[1] < lati and lati < BB_ACT[3] and BB_ACT[0] < longi and longi < BB_ACT[2]):
        return 'ACT'
    elif((BB_NSW1[1] < lati and lati < BB_NSW1[3] and
          BB_NSW1[0] < longi and longi < BB_NSW1[2]) 
          or (BB_NSW2[1] < lati and lati < BB_NSW2[3] 
          and BB_NSW2[0] < longi and longi < BB_NSW2[2])):
        return 'NSW'
    elif(BB_QLD[1] < lati and lati < BB_QLD[3] and BB_QLD[0] < longi and longi < BB_QLD[2]):
        return 'QLD'
    elif(BB_SA[1] < lati and lati < BB_SA[3] and BB_SA[0] < longi and longi < BB_SA[2]):
        return 'SA'
    elif(BB_VIC[1] < lati and lati < BB_VIC[3] and BB_VIC[0] < longi and longi < BB_VIC[2]):
        return 'VIC'
    elif(BB_WA[1] < lati and lati < BB_WA[3] and BB_WA[0] < longi and longi < BB_WA[2]):
        return 'WA'
    elif(BB_TAS[1] < lati and lati < BB_TAS[3] and BB_TAS[0] < longi and longi < BB_TAS[2]):
        return 'TAS'
    elif(BB_NT[1] < lati and lati < BB_NT[3] and BB_NT[0] < longi and longi < BB_NT[2]):
        return 'NT'
    else:
        return 'NOT FOUND'
    
def geolocate(location,lat,lon):
    lat = convert_data_geo(lat)
    lon = convert_data_geo(lon)
    
    if(isinstance(location, basestring)):
        location =location.lower()
        location = remove_special_characters(location)
    else:
        location = 'NO FOUND'
    
    if words_in_string(QUEENSLAND, location) or string_in_another(QUEENSLAND, location):
        return 'QLD'
    elif words_in_string(NSW, location) or string_in_another(NSW, location):
        return 'NSW'
    elif words_in_string(SA, location) or string_in_another(SA, location):
        return 'SA'
    elif words_in_string(TASMANIA, location) or string_in_another(TASMANIA, location):
        return 'TAS'
    elif words_in_string(WA, location) or string_in_another(WA, location):
        return 'WA'
    elif words_in_string(VICTORIA, location) or string_in_another(VICTORIA, location):
        return 'VIC'
    elif words_in_string(NT, location) or string_in_another(NT, location):
        return 'NT'
    elif words_in_string(ACT, location) or string_in_another(ACT, location):
        return 'ACT'
    else:
        if(check_not_none(lat,lon)):
            return find_location(lat,lon)
        else:
            print 'NOT FOUND:',location
            return 'NOT FOUND'

def convert_data_geo(data):
    if(data == 'None'):
        data = None
        return data
    else:
        return float(data)
    
def check_not_none(lat,lon):
    if lat is not None or lon is not None:
        return True
    else:
        return False
    
def get_geodata(df):
    df2 = df[df['PREDICTION']> 0]
    df2['LOCATION'] = df2[['LOCATION','LAT','LON']].apply(lambda x: geolocate(*x), axis=1)
    #df2['LOCATION'] = df2['LOCATION','LAT','LON'].apply(geolocate)
    df2['TIMEZONE'] = pd.DatetimeIndex(df2['TIMEZONE']).normalize()
    return df2[['LOCATION','TIMEZONE']]

def stats_geolocation(df):
    df = get_geodata(df)
    df=df.groupby(['TIMEZONE','LOCATION']).size().reset_index()
    df.columns = ['DATE', 'LOCATION', 'OCCURENCES']
    #df = df.set_index('DATE')
    return df

def get_barchart(df, title):
    df = df.pivot(index= 'DATE', columns='LOCATION', values='OCCURENCES')
    df.fillna(value=0)
    ax = df.plot(kind='bar',stacked=True, title=title,figsize=(9, 7),colormap='Paired')
    # Put a legend to the right of the current axis
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
df_oct = stats_geolocation(geodata_oct)
df_nov = stats_geolocation(geodata_nov)
df_sep = stats_geolocation(geodata_sep)
df = stats_geolocation(geodata)
```
