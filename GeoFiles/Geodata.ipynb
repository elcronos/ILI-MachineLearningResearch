```python
import geopandas
from geopandas.geoseries import *
polygon = geopandas.GeoDataFrame.from_file('shapefiles/STE11aAust.shp') 
#Polygon Australian Regions
nsw_poly = polygon.ix[0]
vic_poly = polygon.ix[1]
qld_poly = polygon.ix[2]
sa_poly = polygon.ix[3]
wa_poly = polygon.ix[4]
tas_poly = polygon.ix[5]
nt_poly = polygon.ix[6]
act_poly = polygon.ix[7]
ot_poly = polygon.ix[8]

def find_location(longi, lati):
    p1 = Point(longi, lati)
    g1= GeoSeries([p1])
    if(g1.intersects(GeoSeries(act_poly.geometry))[0]):
        return 'ACT'
    elif(g1.intersects(GeoSeries(nsw_poly.geometry))[0]):
        return 'NSW'
    elif(g1.intersects(GeoSeries(qld_poly.geometry))[0]):
        return 'QLD'
    elif(g1.intersects(GeoSeries(sa_poly.geometry))[0]):
        return 'SA'
    elif(g1.intersects(GeoSeries(vic_poly.geometry))[0]):
        return 'VIC'
    elif(g1.intersects(GeoSeries(wa_poly.geometry))[0]):
        return 'WA'
    elif(g1.intersects(GeoSeries(tas_poly.geometry))[0]):
        return 'TAS'
    elif(g1.intersects(GeoSeries(nt_poly.geometry))[0]):
        return 'NT'
    elif(g1.intersects(GeoSeries(ot_poly.geometry))[0]):
        return 'OT'
    else:
        return 'NOT FOUND'
```
